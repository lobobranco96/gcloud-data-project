```python
import pandas as pd
import random
import uuid
import os
from faker import Faker
from datetime import datetime, timedelta

fake = Faker('pt_BR')

# Número de registros
NUM_CUSTOMERS = 200_000
NUM_PRODUCTS = 5_000
NUM_ORDERS = 300_000
NUM_DAYS = 30  # Pedidos ao longo de um mês

# Pasta base
today_str = datetime.today().strftime('%Y-%m-%d')
raw_path = "/content/drive/MyDrive/github_projects/gcloud-data-project/raw_data"
raw_dir = f'{raw_path}/{today_str}'
os.makedirs(raw_dir, exist_ok=True)

# Customers
customers = []
existing_customer_ids = []

for _ in range(NUM_CUSTOMERS):
    # 2% chance de duplicar cliente
    if existing_customer_ids and random.random() < 0.02:
        customer_id = random.choice(existing_customer_ids)
    else:
        customer_id = str(uuid.uuid4())
        existing_customer_ids.append(customer_id)

    # 5% chance email faltando
    email = fake.email() if random.random() > 0.05 else None

    # 5% chance created_at futura (erro)
    created_at_date = fake.date_between(start_date='-2y', end_date='today')
    if random.random() < 0.05:
        created_at_date = fake.date_between(start_date='today', end_date='+30d')

    customers.append({
        "customer_id": customer_id,
        "name": fake.name(),
        "email": email,
        "address": fake.address().replace("\n", ", ") + (" " if random.random() < 0.1 else ""),  # espaço extra 10%
        "created_at": created_at_date.strftime('%Y-%m-%d')
    })

df_customers = pd.DataFrame(customers).drop_duplicates(subset=['customer_id'])
df_customers.to_csv(f"{raw_dir}/customers.csv", index=False)


# Products
products = []
existing_product_ids = []
categories = ["Electronics", "Books", "Clothing", "Home", "Beauty", ""]

for _ in range(NUM_PRODUCTS):
    # 2% chance duplicar product_id
    if existing_product_ids and random.random() < 0.02:
        product_id = random.choice(existing_product_ids)
    else:
        product_id = str(uuid.uuid4())
        existing_product_ids.append(product_id)

    # 10% chance categoria inválida ou vazia
    category = random.choice(categories) if random.random() > 0.10 else ""

    # 5% chance preço negativo ou zero
    price = round(random.uniform(10.0, 500.0), 2)
    if random.random() < 0.05:
        price = round(random.uniform(-100.0, 0.0), 2)

    products.append({
        "product_id": product_id,
        "name": fake.word().capitalize(),
        "category": category,
        "price": price,
        "stock": random.randint(0, 1000)
    })

df_products = pd.DataFrame(products).drop_duplicates(subset=['product_id'])
df_products.to_csv(f"{raw_dir}/products.csv", index=False)


# Orders e Itens
orders = []
order_items = []

for i in range(NUM_ORDERS):
    order_id = str(uuid.uuid4())
    customer = random.choice(customers)

    # 5% chance order_date futura (erro)
    order_date = datetime.today() - timedelta(days=random.randint(0, NUM_DAYS - 1))
    if random.random() < 0.05:
        order_date = datetime.today() + timedelta(days=random.randint(1, 10))

    # Status com 10% chance vazio ou inválido
    status = random.choice(["completed", "canceled", "processing", "", "unknown"]) if random.random() > 0.90 else random.choice(["completed", "canceled", "processing"])

    num_items = random.randint(1, 5)

    orders.append({
        "order_id": order_id,
        "customer_id": customer["customer_id"],
        "order_date": order_date.strftime('%Y-%m-%d %H:%M:%S'),
        "status": status
    })

    for _ in range(num_items):
        product = random.choice(products)

        # Quantidade com 2% chance zero ou negativa
        quantity = random.randint(1, 4)
        if random.random() < 0.02:
            quantity = random.choice([0, -1, -2])

        # Unit price pode ser negativo se produto tiver preço negativo
        unit_price = product["price"]

        total_price = round(unit_price * quantity, 2)

        order_items.append({
            "order_id": order_id,
            "product_id": product["product_id"],
            "quantity": quantity,
            "unit_price": unit_price,
            "total_price": total_price
        })

df_orders = pd.DataFrame(orders)
df_order_items = pd.DataFrame(order_items)

df_orders.to_csv(f"{raw_dir}/orders.csv", index=False)
df_order_items.to_csv(f"{raw_dir}/order_items.csv", index=False)


# Inventory updates (simulando entrada/saída de estoque)
inventory_updates = []
for product in products:
    for _ in range(random.randint(1, 3)):
        inventory_updates.append({
            "update_id": str(uuid.uuid4()),
            "product_id": product["product_id"],
            "timestamp": fake.date_time_between(start_date='-30d', end_date='now').strftime('%Y-%m-%d %H:%M:%S'),
            "change": random.choice([-10, -5, 10, 20, 50])
        })

df_inventory = pd.DataFrame(inventory_updates)
df_inventory.to_csv(f"{raw_dir}/inventory_updates.csv", index=False)

print(f"Dados gerados em: {raw_dir}")
```

```python
import pandas as pd
import random
import uuid
import os
from faker import Faker
from datetime import datetime, timedelta

fake = Faker('pt_BR')

# Número de registros
NUM_CUSTOMERS = 200_000
NUM_PRODUCTS = 5_000
NUM_ORDERS = 300_000
NUM_DAYS = 30  # Pedidos ao longo de um mês

# Pasta base
today_str = datetime.today().strftime('%Y-%m-%d')
raw_path = "/content/drive/MyDrive/github_projects/gcloud-data-project/raw_data"
raw_dir = f'{raw_path}/{today_str}'
os.makedirs(raw_dir, exist_ok=True)

# Customers
customers = []
existing_customer_ids = []

for _ in range(NUM_CUSTOMERS):
    # 2% chance de duplicar cliente
    if existing_customer_ids and random.random() < 0.02:
        customer_id = random.choice(existing_customer_ids)
    else:
        customer_id = str(uuid.uuid4())
        existing_customer_ids.append(customer_id)

    # 5% chance email faltando
    email = fake.email() if random.random() > 0.05 else None

    # 5% chance created_at futura (erro)
    created_at_date = fake.date_between(start_date='-2y', end_date='today')
    if random.random() < 0.05:
        created_at_date = fake.date_between(start_date='today', end_date='+30d')

    customers.append({
        "customer_id": customer_id,
        "name": fake.name(),
        "email": email,
        "address": fake.address().replace("\n", ", ") + (" " if random.random() < 0.1 else ""),  # espaço extra 10%
        "created_at": created_at_date.strftime('%Y-%m-%d')
    })

df_customers = pd.DataFrame(customers).drop_duplicates(subset=['customer_id'])
df_customers.to_csv(f"{raw_dir}/customers.csv", index=False)


# Products
products = []
existing_product_ids = []
categories = ["Electronics", "Books", "Clothing", "Home", "Beauty", ""]

for _ in range(NUM_PRODUCTS):
    # 2% chance duplicar product_id
    if existing_product_ids and random.random() < 0.02:
        product_id = random.choice(existing_product_ids)
    else:
        product_id = str(uuid.uuid4())
        existing_product_ids.append(product_id)

    # 10% chance categoria inválida ou vazia
    category = random.choice(categories) if random.random() > 0.10 else ""

    # 5% chance preço negativo ou zero
    price = round(random.uniform(10.0, 500.0), 2)
    if random.random() < 0.05:
        price = round(random.uniform(-100.0, 0.0), 2)

    products.append({
        "product_id": product_id,
        "name": fake.word().capitalize(),
        "category": category,
        "price": price,
        "stock": random.randint(0, 1000)
    })

df_products = pd.DataFrame(products).drop_duplicates(subset=['product_id'])
df_products.to_csv(f"{raw_dir}/products.csv", index=False)


# Orders e Itens
orders = []
order_items = []

for i in range(NUM_ORDERS):
    order_id = str(uuid.uuid4())
    customer = random.choice(customers)

    # 5% chance order_date futura (erro)
    order_date = datetime.today() - timedelta(days=random.randint(0, NUM_DAYS - 1))
    if random.random() < 0.05:
        order_date = datetime.today() + timedelta(days=random.randint(1, 10))

    # Status com 10% chance vazio ou inválido
    status = random.choice(["completed", "canceled", "processing", "", "unknown"]) if random.random() > 0.90 else random.choice(["completed", "canceled", "processing"])

    num_items = random.randint(1, 5)

    orders.append({
        "order_id": order_id,
        "customer_id": customer["customer_id"],
        "order_date": order_date.strftime('%Y-%m-%d %H:%M:%S'),
        "status": status
    })

    for _ in range(num_items):
        product = random.choice(products)

        # Quantidade com 2% chance zero ou negativa
        quantity = random.randint(1, 4)
        if random.random() < 0.02:
            quantity = random.choice([0, -1, -2])

        # Unit price pode ser negativo se produto tiver preço negativo
        unit_price = product["price"]

        total_price = round(unit_price * quantity, 2)

        order_items.append({
            "order_id": order_id,
            "product_id": product["product_id"],
            "quantity": quantity,
            "unit_price": unit_price,
            "total_price": total_price
        })

df_orders = pd.DataFrame(orders)
df_order_items = pd.DataFrame(order_items)

df_orders.to_csv(f"{raw_dir}/orders.csv", index=False)
df_order_items.to_csv(f"{raw_dir}/order_items.csv", index=False)


# Inventory updates (simulando entrada/saída de estoque)
inventory_updates = []
for product in products:
    for _ in range(random.randint(1, 3)):
        inventory_updates.append({
            "update_id": str(uuid.uuid4()),
            "product_id": product["product_id"],
            "timestamp": fake.date_time_between(start_date='-30d', end_date='now').strftime('%Y-%m-%d %H:%M:%S'),
            "change": random.choice([-10, -5, 10, 20, 50])
        })

df_inventory = pd.DataFrame(inventory_updates)
df_inventory.to_csv(f"{raw_dir}/inventory_updates.csv", index=False)

print(f"Dados gerados em: {raw_dir}")
```
